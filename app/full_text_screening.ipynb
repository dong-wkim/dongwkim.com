```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd

#url = "../../../systematic_review/screening/full_text_screening/full_text_screening.csv"
url = "data/full_text_screening.csv"

df = pd.read_csv(url, encoding = "utf-8")
df["full_text_screening"] = ""
df["reason_for_exclusion"] = ""

a = widgets.IntText(value = 0, description = "ID ", min = 1, max = len(df)-1, layout = {"width": "50%"})
a3 = widgets.HTML(value = "/" + str(len(df)))
b = widgets.Text(value = "", description = "study ", layout = {"width": "50%"}) # study
c = widgets.Dropdown(options = ['BPTB', 'HT', 'QT', 'PLT', 'AT', 'TA'], value = "BPTB", description = "subgroup", layout = {"width": "50%"})
d = widgets.Text(value = "", description = "author(s) ", layout = {"width": "90%"}) # authors
e = widgets.Text(value = "", description = "title ", layout = {"width": "90%"}) # title
f = widgets.HTML(value = "", description = "abstract ", layout = {"width": "91.5%", "height": "300%"}) # abstract
g = widgets.Text(value = "", description = "year ", layout = {"width": "90%"}) # year
h = widgets.Text(value = "", description = "journal ", layout = {"width": "90%"}) # journal
i = widgets.Dropdown(options = ["PubMed", "Embase", "Web of Science"], value = "PubMed", description = "source", layout = {"width": "50%"})
j = widgets.Text(value = "", description = "DOI ", layout = {"width": "90%"}) # doi
k = widgets.Text(value = "", description = "URL ", layout = {"width": "90%"}) # url
l = widgets.Text(placeholder = "Reason for exclusion")
    
def update(change):
    study = df.loc[df["id"] == a.value, "study"]
    subgroup =  df.loc[df["id"] == a.value, "subgroup"]
    authors =  df.loc[df["id"] == a.value, "authors"]
    title =  df.loc[df["id"] == a.value, "title"]
    abstract =  df.loc[df["id"] == a.value, "abstract"]
    year =  df.loc[df["id"] == a.value, "year"]
    journal =  df.loc[df["id"] == a.value, "journal"]
    source =  df.loc[df["id"] == a.value, "source"]
    doi =  df.loc[df["id"] == a.value, "doi"]
    url =  df.loc[df["id"] == a.value, "doi_url"]
    reason =  df.loc[df["id"] == a.value, "reason_for_exclusion"]

    if a.value > 0:
       b.value = str(study.iloc[0])
       d.value = str(authors.iloc[0])
       e.value = str(title.iloc[0])
       f.value = str(abstract.iloc[0])
       g.value = str(year.iloc[0])
       h.value = str(journal.iloc[0])
       j.value = str(doi.iloc[0])
       k.value = str(url.iloc[0])
       l.value = str(reason.iloc[0])
        
    else:
        b.value = ""
        d.value = ""
        e.value = ""
        f.value = ""
        g.value = ""
        h.value = ""
        j.value = ""
        k.value = ""
        l.value = ""

a.observe(update, names = "value") # observe widget 'a' (i.e., ID column) for any changes, and if there is, then display all of the widgets corresponding to the value of 'a'.

included = widgets.Button(value = "Included", description = "Included", style = {"button_color":"green"})
def included_click(button):
    df.loc[df["id"] == a.value, "full_text_screening"] = "included"
included.on_click(included_click)

# Excluded reports from full-text screening
excluded = widgets.Button(value = "Excluded", description = "Excluded", style = {"button_color":"red"})
def excluded_click(button):
    df.loc[df["id"] == a.value, "full_text_screening"] = "excluded"
excluded.on_click(excluded_click)

# PRISMA Reason for exclusion 
submit = widgets.Button(description = "Submit")
reason_submit = widgets.HBox(children = [l, submit])

def reason_submit_click(button):
    df.loc[df["id"] == a.value, "reason_for_exclusion"] = str(l.value)
    
submit.on_click(reason_submit_click)

def save_click(button):
    import os
    os.makedirs("./data", exist_ok = True)
    df.to_csv("./data/full_text_screening.csv", encoding = "utf-8")
    
save = widgets.Button(value = "Save", description = "Save", style = {"button_color":"blue"})
save.on_click(save_click)
#toggle_buttons = widgets.ToggleButtons(options = ["Yes", "No", "Maybe"])
out = widgets.Output()

title = widgets.HTML(value = "<h1>Full-text screening form<h1>")
A = widgets.VBox(children = [widgets.HBox(children = [a, a3]), b, c, d], layout = {"width": "50%"})
B = widgets.VBox(children = [included, excluded, reason_submit, save], layout = {"width": "50%"})
AB = widgets.HBox(children = [A, B], layout = {"width": "100%"})
C = widgets.VBox(children = [e, f, g, h, i , j, k], layout = {"width": "100%"})

full_text_screening = widgets.VBox(children = [title, AB, C], layout = {"width": "100%"})
display(full_text_screening, out)
```
